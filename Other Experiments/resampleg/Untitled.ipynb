```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV files
df_resample = pd.read_csv("outputs/training_log_resample_seed_67.csv")
df_regular = pd.read_csv("outputs/training_log_regular_seed_67.csv")

# Plot training loss
plt.figure(figsize=(8, 5))
plt.plot(df_resample["epoch"], df_resample["train_loss"], linewidth=2, label="resample")
plt.plot(df_regular["epoch"], df_regular["train_loss"], linewidth=2, label="regular")

plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.title("Training Loss Comparison")
plt.legend(frameon=False, loc="upper right")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
